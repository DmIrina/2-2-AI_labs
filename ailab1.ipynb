```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 16, 0.1)
y = fuzz.trimf(x,[4, 8, 12])
plt.figure(figsize=(10, 5))
plt.title('Трикутна функція приналежності, P = [4, 8, 12]')
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x,y)
```

```python
x = np.arange(0, 16, 0.1)
y = fuzz.trapmf(x,[6, 8, 10, 12])
plt.figure(figsize=(10, 5))
plt.title('Трапецієподібна функція приналежності, P = [6, 8, 10, 12]')
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x,y)
```

```python
x = np.arange(-20, 30, 0.1)
y = fuzz.gaussmf(x, 7, 8)
plt.figure(figsize=(10, 5))
plt.title('Проста функція приналежності Гаусса')
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x,y)
```

```python
x = np.arange(-30, 30, 0.1)
y = fuzz.gauss2mf(x, -7, 10, 0, 15)
plt.figure(figsize=(25, 10))
plt.title('Двостороння функція приналежності Гаусса')
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x,y)
```

```python
x = np.arange(-20, 30, 0.1)
y1 = fuzz.gaussmf(x, 7, 8)
y2 = fuzz.gaussmf(x, -7, 8)
minimum = np.minimum(y1, y2)
plt.figure(figsize=(20, 15))
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x, y1, linestyle = '--')
plt.plot(x, y2, linestyle = '--')
plt.plot(x, minimum)
```

```python
x = np.arange(-20, 30, 0.1)
y1 = fuzz.gaussmf(x, 7, 8)
y2 = fuzz.gaussmf(x, -7, 8)
maximum = np.maximum(y1, y2)
plt.figure(figsize=(20, 15))
plt.xlabel('x')
plt.ylabel('Рівень приналежності')
plt.plot(x, y1, linestyle = '--')
plt.plot(x, y2, linestyle = '--')
plt.plot(x, maximum)
```

```python
x = np.arange(-20, 30, 0.1)
y1 = fuzz.gaussmf(x, 7, 8)
y2 = fuzz.gaussmf(x, -7, 8)
prod = y1 * y2
plt.figure(figsize=(20, 15))
plt.title("Кон'юнкція")
plt.plot(x, y1, linestyle = '--')
plt.plot(x, y2, linestyle = '--')
plt.plot(x, prod)
```

```python
x = np.arange(-20, 30, 0.1)
y1 = fuzz.gaussmf(x, 7, 8)
y2 = fuzz.gaussmf(x, -7, 8)
probor = (y1 + y2 - y1 * y2) / np.max(y1 + y2 - y1 * y2)
plt.figure(figsize=(20, 15))
plt.title("Диз'юнкція")
plt.plot(x, y1, linestyle = '--')
plt.plot(x, y2, linestyle = '--')
plt.plot(x, probor)
```

```python
x = np.arange(-20, 30, 0.1)
y1 = fuzz.gaussmf(x, 7, 8)
y2 = 1 - y1
plt.figure(figsize=(20, 15))
plt.title("Доповнення")
plt.plot(x, y1, linestyle = '--')
plt.plot(x, y2)
```
