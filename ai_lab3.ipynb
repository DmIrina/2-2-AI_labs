```python
import skfuzzy as fuzz
import numpy as np
import matplotlib.pyplot as plt

from sklearn import datasets
```

```python
n_samples = 1000
xy, labels = datasets.make_blobs(n_samples=n_samples, random_state=50)
x, y = xy.T
```

```python
plt.figure(figsize=(20, 15))
for label in range(len(np.unique(labels))):
    plt.plot(x[labels == label], y[labels == label], 'x')
plt.show()
```

```python
plt.figure(figsize=(20, 15))
plt.plot(x, y, 'x')
plt.show()
```

```python
fpcs = []

n = 10
for c in range(2, 10):
    cntr, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(
        np.array([x, y]), c, 2, error=0.005, maxiter=1000, init=None)
    fpcs.append(fpc)
```

```python
print(fpcs)
```

```python
plt.figure(figsize=(20, 15))
plt.plot(np.arange(2, 10), fpcs, color='#731810')
plt.title("How Number of Clusters Change FPC?")
plt.xlabel("Number of centers")
plt.ylabel("Fuzzy partition coefficient")
plt.show()
```

```python
n_centers = 2
```

```python
centers, u, u0, d, jm, p, fpc = fuzz.cluster.cmeans(np.array([x, y]), n_centers, 2, error=0.005, maxiter=1000, init=None)
```

```python
cluster_membership = np.argmax(u, axis=0)
```

```python
plt.figure(figsize=(20, 15))
plt.title('fps = ' + str(fpc))

for j, c in enumerate(centers):
    plt.plot(x[cluster_membership == j], y[cluster_membership == j], 'x')
    plt.plot(c[0], c[1], 'X', color = 'cyan', ms = 10)

plt.show()
```
